#### Linear Regression using sklearn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, r2_score
```

```python
# student data
study_hours = np.array([1, 2, 2.5, 3, 4, 4.5, 5, 6, 6.5, 7, 8, 9, 10]) 
marks_scored = np.array([20, 40, 50, 55, 60, 65, 70, 75, 78, 80, 85, 90, 95])  
```

```python
df = pd.DataFrame({'Study Hours': study_hours, 'Marks Scored': marks_scored})

df.head()
```

```python

X = df[['Study Hours']]  # independent var
Y = df['Marks Scored']  # dependent var

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(f"Training samples: {len(X_train)}, Testing samples: {len(X_test)}")
```

##### Implementing linear regression 

```python
# implemting Linear Regression model
model = LinearRegression()

# Training model
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
```

##### Calculating Evaluation Metrices

```python
mse = mean_squared_error(Y_test, Y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(Y_test, Y_pred)


print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

##### Data Visualization

```python

plt.scatter(X_test, Y_test, color='blue', label="Actual values")  
plt.plot(X_test, Y_pred, color='red', linestyle="dashed", label="Predicted values") 
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Study Hours vs Marks Scored (Linear Regression)")
plt.legend()
```
