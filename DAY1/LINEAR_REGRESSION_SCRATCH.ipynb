```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# data values of students where X is study hours and Y is marks scored

X = np.array([5,6,7,8,9,10,4,3,7,5,2,9,4,2,3,3 ,7,6])  # Study Hours
Y = np.array([78,87,80,89,90,34,33,76,50,20,88,76,23,43,31,66,88,30])  # Exam Scores
```

```python
# COST_FUNCTION(OR MEAN SQUARED ERROR)
# Cost Function measures how far predictions are from actual values


def cost_func(m, b, X, Y):
    n = len(X) 
    total_error = 0

    for i in range(n):
        predicted = m * X[i] + b 
        total_error += (predicted - Y[i]) ** 2  

    return total_error / (2 * n) 
```

```python
# IMPLEMENTING GRADIENT DESCENT
#Gradient Descent optimizes m (slope) and b (intercept) to minimize error

def gradient_descent(X, Y, m, b, learning_rate, epochs):
    n = len(X) 

    for _ in range(epochs):
        m_gradient = 0
        b_gradient = 0

        for i in range(n):
            predicted = m * X[i] + b  
            error = predicted - Y[i]

            m_gradient += error * X[i] 
            b_gradient += error  

        m -= (learning_rate * m_gradient) / n  
        b -= (learning_rate * b_gradient) / n  

    return m, b
```

```python
# TRAIN THE MODEL

m = 0  # Initial slope
b = 0  # Initial intercept
learning_rate = 0.01
epochs = 1000

m, b = gradient_descent(X, Y, m, b, learning_rate, epochs)

print(f"Optimized Slope (m): {m}")
print(f"Optimized Intercept (b): {b}")

```

```python
# MAKE PREDICTIONS AND COMPARE

predictions = m * X + b

# Plotting"
plt.scatter(X, Y, color='red', label="Actual Data")
plt.plot(X, predictions, color='blue', label="Predicted Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.title("Linear Regression from Scratch")

```
